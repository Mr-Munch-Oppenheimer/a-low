```python
import statsmodels.api as sm
```

```python
# 去极值
def winsorize(factor, std=3):
    r = factor.copy()
    r_mean = r.mean()
    r_std = r.std()
    edge_up = r_mean + std * r_std
    edge_low = r_mean - std * r_std
    r[r>edge_up]=edge_up
    r[r<edge_low]=edge_low
    return r
```

```python
# 标准化
def standardize(factor, ty=2):
    r = factor.copy()
    if ty==1:
        r = (r-r.min())/(r.max()-r.min())
    elif ty==2:
        r = (r-r.mean())/r.std()
#     elif ty==3:
#         r= r/10**np.ceil(np.log10(r.abs().max()))
    return r
```

```python
# 中性化
def neutral(x,y,industry = True,mkt_cap = False):
    result = sm.WLS(y.astype(float), x.astype(float)).fit()
#     print(y)
#     print(x)
    return result.resid
```

```python
zz500=pd.read_csv("zz500.csv")

@joblib.delayed
def task(current_date,i):
    
    current_df =pd.DataFrame({"factor":factor_df.loc[current_date],"return":return_df.loc[current_date]})
    current_df.dropna(inplace=True)
    if current_df.shape[0]==0:
        return pd.DataFrame()
    
    current_df["factor"]=winsorize(current_df.factor)
    current_df["factor"]=standardize(current_df.factor)

    if industry_neutral == True:
        industry_df = pd.read_csv("industry/"+str(current_date)+".csv")
        industry_df.set_index("ts_code",inplace=True)
        new_industry_df = pd.DataFrame(index= current_df.index.tolist(),columns=industry_df.columns)
        for col in industry_df.columns:
            new_industry_df[col]=industry_df[col]

    if lnMarketCap_neutral == True:
        lnMarketCap_df=pd.read_csv("lnMarketCap/"+str(current_date)+".csv")
        lnMarketCap_df.set_index("ts_code",inplace=True)
        new_industry_df["lnMarketCap"]=lnMarketCap_df["lnMarketCap"].copy()

    if industry_neutral == True:
        current_df["factor"]=neutral(new_industry_df,current_df.factor)
    
    current_df.sort_values(by="factor",inplace=True)#从小到大排序
    
    single_group_num = int(current_df.shape[0]/10)
    
    group01 = current_df["return"][0:single_group_num].mean()
    group02 = current_df["return"][single_group_num:2*single_group_num].mean()
    group03 = current_df["return"][2*single_group_num:3*single_group_num].mean()
    group04 = current_df["return"][3*single_group_num:4*single_group_num].mean()
    group05 = current_df["return"][4*single_group_num:5*single_group_num].mean()
    group06 = current_df["return"][5*single_group_num:6*single_group_num].mean()
    group07 = current_df["return"][6*single_group_num:7*single_group_num].mean()
    group08 = current_df["return"][7*single_group_num:8*single_group_num].mean()
    group09 = current_df["return"][8*single_group_num:9*single_group_num].mean()
    group10 = current_df["return"][9*single_group_num:].mean()
    
#     return_mean_stock = current_df["return"][0:first_num].mean()
    corr_spearman = current_df.corr(method='spearman').iloc[0][1]
    info_df=pd.DataFrame({"group01":group01,
                          "group02":group02,
                          "group03":group03,
                          "group04":group04,
                          "group05":group05,
                          "group06":group06,
                          "group07":group07,
                          "group08":group08,
                          "group09":group09,
                          "group10":group10,
                          "corr_spearman":corr_spearman,
                          "zz500_future_5":zz500.iloc[i].future_5,
                          "current_date":current_date},index=[current_date])
    return info_df

tasks=[]
i=0
while i<trade_cal.shape[0]:
#     print(i,trade_cal.shape[0])
#     if i<2400:
#         i=i+step
#         continue
    current_date = trade_cal.cal_date[i]
    tasks.append(task(current_date,i))
    i=i+step    
    
df = pd.concat(
    joblib.Parallel(n_jobs=5, verbose=10)(
        tasks
    ), axis=0)
```
