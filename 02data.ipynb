```python
custom_date = pd.read_csv("custom_date.csv")
start_date=str(custom_date.start_date[0])
end_date=str(custom_date.end_date[0])
print(start_date)
print(end_date)
```

```python
if not os.path.isdir("stock_pool"):
    os.makedirs("stock_pool")
    print("已创建stock_pool文件夹")
else:
    print("stock_pool文件夹已存在")
```

```python
all_index = stock_basic.shape[0]
df_index = pd.to_datetime(trade_cal.cal_date.tolist(),format='%Y%m%d')
```

```python
#停牌
path = "stock_pool/suspend"
if not os.path.isdir(path):
    os.makedirs(path)
    print("已创建"+path+"文件夹")
else:
    print(path+"文件夹已存在")
for index, row in stock_basic.iterrows():
    print(index,stock_basic.shape[0])
    stock_name=row.ts_code
    current_df = pro.suspend_d(suspend_type='S', ts_code=stock_name)
    current_df.sort_values(by='trade_date',ascending=True,inplace=True)
    current_df.reset_index(inplace=True,drop=True)
    current_df.trade_date=pd.to_datetime(current_df.trade_date.tolist(),format='%Y%m%d')
    current_df.index=current_df.trade_date
    standard_df = pd.DataFrame(index=df_index,columns=current_df.columns.tolist())
    for column in current_df.columns:
        standard_df[column]=current_df[column]
    standard_df.trade_date=standard_df.index.tolist()
    standard_df.ts_code = stock_name
    standard_df.to_csv(path+"/"+stock_name+".csv",index=0)
    time.sleep(0.12)
# for index, row in trade_cal.iterrows():
#     print(index)
#     df = pro.suspend_d(suspend_type='S', trade_date=row.cal_date)
#     df.to_csv(path+"/"+str(row.cal_date)+".csv",index=0)
```

```python
#涨跌停
path = "stock_pool/limit"
if not os.path.isdir(path):
    os.makedirs(path)
    print("已创建"+path+"文件夹")
else:
    print(path+"文件夹已存在")
for index, row in stock_basic.iterrows():
    print(index,stock_basic.shape[0])
    stock_name=row.ts_code
    current_df = pro.stk_limit(ts_code=stock_name,start_date=start_date, end_date=end_date)    
    current_df.sort_values(by='trade_date',ascending=True,inplace=True)
    current_df.reset_index(inplace=True,drop=True)
    current_df.trade_date=pd.to_datetime(current_df.trade_date.tolist(),format='%Y%m%d')
    current_df.index=current_df.trade_date
    standard_df = pd.DataFrame(index=df_index,columns=current_df.columns.tolist())
    for column in current_df.columns:
        standard_df[column]=current_df[column]
    standard_df.trade_date=standard_df.index.tolist()
    standard_df.ts_code = stock_name
    standard_df.to_csv(path+"/"+stock_name+".csv",index=0)
    time.sleep(0.12)
```

```python
#ST
path = "stock_pool/st"
if not os.path.isdir(path):
    os.makedirs(path)
    print("已创建"+path+"文件夹")
else:
    print(path+"文件夹已存在")
for index, row in stock_basic.iterrows():
    print(index,stock_basic.shape[0])
    stock_name=row.ts_code
    df = pro.namechange(ts_code=stock_name, fields='ts_code,name,start_date,end_date,change_reason')
    df.to_csv(path+"/"+stock_name+".csv",index=0)
    time.sleep(0.12)
```
