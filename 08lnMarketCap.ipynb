```python
import pandas as pd
import numpy as np
```

```python
trade_cal = pd.read_csv("trade_cal.csv")
trade_cal
```

```python
factor_df=pd.read_csv("total_mv.csv")
factor_df
```

```python
factor_df=np.log(factor_df)
factor_df.fillna(0, inplace=True)
factor_df
```

```python
for i in range(0,trade_cal.shape[0]):
    current_date = str(trade_cal.cal_date[i])
    print(i,trade_cal.shape[0],current_date)
    df = pd.DataFrame(index=factor_df.columns,columns=["lnMarketCap","ts_code"])
    df.lnMarketCap = factor_df.iloc[i]
    df.ts_code = factor_df.columns.tolist()
    df.to_csv("lnMarketCap/"+current_date+".csv",index=0)
```
