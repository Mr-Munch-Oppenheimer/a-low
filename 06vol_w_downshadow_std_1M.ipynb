```python
path = "factor/vol_w_downshadow_std_1M"
```

```python
if not os.path.isdir(path):
    os.makedirs(path)
    print("已创建"+path+"文件夹")
else:
    print(path+"文件夹已存在")
```

```python
current_df=pd.read_csv("stock_daily/000001.SZ.csv")
current_df=current_df[["ts_code","trade_date","low","close"]]
current_df["w"]=(current_df.close-current_df.low)/current_df.low
current_df["factor"]=current_df["w"].rolling(window=20).std()
current_df
```

```python
@joblib.delayed
def task(index):
    ts_code = stock_basic.ts_code[index]
    current_df=pd.read_csv("stock_daily/"+ts_code+".csv")
    current_df=current_df[["ts_code","trade_date","open","close"]]
    current_df["intraday"]=(current_df.close-current_df.open)/current_df.open
    current_df["factor"]=current_df["intraday"].rolling(window=20).sum()
    current_df.to_csv(path+"/"+ts_code+".csv",index=0)
    return index
tasks=[]
for j in range(stock_basic.shape[0]):
#     print("stock_daily/"+stock_basic.ts_code[i]+".csv")
    tasks.append(task(j))
parallel = joblib.Parallel(n_jobs=5, verbose=10)
parallel(tasks)
```
